```
import csv
import pandas as pd
```

```
inch_lists = [
[' 11 ','11형','11인치','11.'],
[' 12 ','12형','12인치','12.'],
[' 13 ','13형','13인치','13.'],
[' 15 ','15형','15인치','15.'],
[' 16 ','16형','16인치','16.']]

displays = ["11인치(27~29cm)","12인치(29~31cm)","13인치(32~35cm)","15인치(37~39cm)","16인치(40~42cm)"]
inch_size = len(inch_lists)
each_inch_size = len(inch_lists[0])
print('inch_size : ', inch_size)
print('each_inch_size : ', each_inch_size)
for i in range(inch_size):
  print(displays[i])
```

```
def getDisplay(name):
  for i in range(inch_size): # 11인치 ~ 16인치 배열로 비교하자
    for j in range(each_inch_size): # 각 인치당 str size만큼
      result = name.find(inch_lists[i][j])
      if result > -1: # 찾는 문자열 O
        return displays[i]
  return 'missing'
```

```
# 빈 칸 Display 데이터 채우기 전처리
inp = open('/content/drive/MyDrive/SK Python/2021.01 프로그래머스 데이터분석/train_display_final.csv','r',encoding='CP949')
out = open('/content/drive/MyDrive/SK Python/2021.01 프로그래머스 데이터분석/train_display_final2.csv','w',encoding='CP949')

inp_read = csv.reader(inp)
out_write = csv.writer(out)
idx = -1
for line in inp_read:
  name = line[1]
  display = line[2]
  print(name)
  if display == '':
    display = getDisplay(name)
  print('display : ',display)
  out_write.writerow([idx, name, display])
  idx=idx+1
inp.close()
out.close()
```
